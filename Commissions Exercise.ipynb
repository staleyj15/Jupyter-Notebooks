```python
import pandas as pd
from itertools import product
from math import floor
import matplotlib.pyplot as plt
import holoviews as hv
import seaborn as sns
hv.extension('bokeh')

# Import commissions data into dataframe
sales = pd.read_csv('/mnt/c/Users/jessi/OneDrive/Documents/Python_Practice/Toast/Commission Case Study.csv', sep = ',', header = 0, names = ['opp_id', 'opp_name', 'restaurant_type', 'opp_owner', 'close_date', 'live_date', 'churn_date', 'deposit_received', 'saas_rev', 'payment_rev', 'hardware_val', 'service_val'])

# Change column data types
sales['deposit_received'] = sales['deposit_received'].apply(lambda x: True if x == 'Yes' else False)
sales['close_date'] = pd.to_datetime(sales['close_date'])
sales['live_date'] = pd.to_datetime(sales['live_date'])
sales['churn_date'] = pd.to_datetime(sales['churn_date'])
for col in ['saas_rev', 'payment_rev', 'hardware_val', 'service_val']:
    sales[col] = sales[col].str.strip(' $').str.replace(",", "").astype('int32')
    
# Define functions
def generate_barvals_ccv(df, owner): 
    '''
    Generate values for stacked bar chart
    '''    
    index = list(df.loc[df['opp_owner'] == owner, 'opp_name'].values)
    groups = ['saas_rev','payment_rev','hardware_val','service_val']
    vals = []
    
    for g in groups:
        t = []
        for i in index:
            if g in ['saas_rev','payment_rev']: t.append(df.loc[(df['opp_owner'] == owner) & (df['opp_name'] == i), g].values[0] * 12)
            else: t.append(df.loc[(df['opp_owner'] == owner) & (df['opp_name'] == i), g].values[0] * .5)
        vals.append(t)

    return vals

def plot_ccv_bars(df, names):
    '''
    Generate stacked bar chart showing distribution of opportunity ccv by contributing revenue
    '''
    fig, ax = plt.subplots(2, 2, figsize=(26,15))
    ax = ax.ravel()
    fig.tight_layout()
    colors = ['#f7546f','#4dd0e1','#ff7043','#a3cf77']
    
    for i, name in enumerate(names):
        vals = generate_barvals_ccv(df, name)
        N = len(vals[0])
        ind = np.arange(N)
        width = .2

        for ii in range(len(vals)):
            ax[i].bar(ind + width * ii, vals[ii], width, color = colors[ii])
            
        ax[i].spines['right'].set_visible(False)
        ax[i].spines['top'].set_visible(False)
        ax[i].set_title(' '.join(['Distribution of CCV for', name, 'by Opportunity Name']))
        ax[i].set_xlabel('Opportunity Name')
        ax[i].set_ylabel('Dollar Contribution')
        ax[i].set_xticks(ind + width)
        ax[i].set_xticklabels(list(df.loc[df['opp_owner'] == name, 'opp_name'].values), rotation=50, ha='right')
        ax[i].legend(('Software Revenue','Payment Revenue', 'Hardware Revenue', 'Services Revenue'))
    
    plt.subplots_adjust(hspace = 0.4)

def generate_barvals(df, owner):
    '''
    Generate values for Bokeh stacked bar chart
    '''
    
    index = list(df.loc[df['opp_owner'] == owner, 'opp_name'].values)
    groups = ['Recovered for Churn', 'Deposit Kicker', 'CCV Payment 1', 'CCV Payment 2']
    pos = []
    neg = []
    
    for i in index:
        ix4 = [i for ix in range(4)]
        
        v1 = floor(df.loc[(df['opp_owner'] == owner) & (df['opp_name'] == i), 'recovered'].values[0])
        v2 = floor(df.loc[(df['opp_owner'] == owner) & (df['opp_name'] == i), 'kicker'].values[0])
        v3 = floor(df.loc[(df['opp_owner'] == owner) & (df['opp_name'] == i), 'net_p1'].values[0])
        v4 = floor(df.loc[(df['opp_owner'] == owner) & (df['opp_name'] == i), 'gross_p2'].values[0])

        tup = (0, v2, v3, v4)
        pos.append(list(zip(ix4, groups, tup)))
        neg.append([(i, v1)])
        
    pos = [item for sublist in pos for item in sublist]
    neg = [item for sublist in neg for item in sublist]
    
    return pos, neg

def plot_pmtsByOpp(df, owner):
    '''
    Generate stacked bar plot of commission payments per opportunity name and owner
    '''
    
    pos, neg = generate_barvals(df, owner)
    minN = min([b for a,b in neg])
    
    pos_stacked = hv.Bars(pos, ['Opportunity Name', 'Payment Type'], 'Commission Dollars').opts(stacked = True, clone = True, width = 900, height = 450, tools = ['hover'], title = ''.join(['Commission Dollars by Opportunity for ', owner]), xlabel = '', ylim = (0,1150))
    neg_bar = hv.Bars(neg, ['Opportunity Name'], 'Commission Dollars').opts(width = 900, height = 220, tools = ['hover'], xaxis = None, yticks = [-500,-400,-300,-200,-100], ylabel = '', ylim=(-450,0))
    layout = pos_stacked + neg_bar
    
    if minN == 0: return pos_stacked
    else: return layout.opts(shared_axes = False).cols(1)
    
# Define functions
def generate_ccv_pct(df, owner): 
    '''
    Generate values for stacked bar chart
    '''    
    index = list(df.loc[df['opp_owner'] == owner, 'opp_name'].values)
    groups = [calc_ccv, calc_commission]
    vals = [[1 for i in range(len(index))]]
    t = []
    
    for i in index:
        df_sub = df.loc[(df['opp_owner'] == owner) & (df['opp_name'] == i)].copy()
        ccv = df_sub.apply(calc_ccv, axis = 1).values[0]
        comm = df_sub.apply(calc_commission, axis = 1).values[0]
        t.append(comm / ccv)
    
    vals.append(t)
    
    return vals

def plot_ccv_pct(df, names):
    '''
    Generate stacked bar chart showing percent of CCV sales representative gains commission on
    '''
    fig, ax = plt.subplots(1, 3, figsize=(26, 8))
    ax = ax.ravel()
    fig.tight_layout()
    colors = ['#2583e3','#ff4c00']
    
    for i, name in enumerate(names):
        vals = generate_ccv_pct(df, name)
        N = len(vals[0])
        ind = np.arange(N)
        width = .4

        for ii in range(len(vals)):
            ax[i].bar(ind + width * ii, vals[ii], width, color = colors[ii])
            
        ax[i].spines['right'].set_visible(False)
        ax[i].spines['top'].set_visible(False)
        ax[i].set_title(' '.join(['CCV vs Total Commission for', name, 'by Opportunity Name']))
        ax[i].set_xlabel('Opportunity Name')
        ax[i].set_ylabel('Percent of CCV')
        ax[i].set_xticks(ind + width)
        ax[i].set_xticklabels(list(df.loc[df['opp_owner'] == name, 'opp_name'].values), rotation=50, ha='right')
        ax[i].legend(('CCV','Total Commission'))
    
    plt.subplots_adjust(hspace = 0.4)

def pay_details(df):
    '''
    Calculate payments by date
    '''
    
    start = dt.strptime('2019-06-01', '%Y-%m-%d')
    end = dt.strptime('2019-06-30', '%Y-%m-%d')
    if df['close_date'] >= start and df['close_date'] <= end: 
        return df['close_date'], df['gross_p1']
    else: 
        return df['live_date'], df['gross_p2']
    
def plot_pmtbyDay(df, names, cumulative = True):
    '''
    Plot commission payments by date
    '''
    
    yval = 'pmt_cumsum' if cumulative == True else 'pmt_amt'
    ylab = 'Cumulative Payments' if cumulative == True else 'Payment Amount'
    x_dates = df['pmt_dt'].dt.strftime('%m-%d').unique()
    colors = ['#f7546f','#4dd0e1','#ff7043','#a3cf77']
    
    fig, ax = plt.subplots(2, 2, figsize = (26,15))
    ax = ax.ravel()

    for i, name in enumerate(names):
        title = ''.join(['Cumulative Commission Payments for ', name]) if cumulative == True else ''.join(['Commission Payments for ', name])
        sns.barplot(x = 'pmt_dt', y = yval, data = df.loc[df['opp_owner'] == name,], ci = None, ax = ax[i], color = colors[i])
        ax[i].spines['right'].set_visible(False)
        ax[i].spines['top'].set_visible(False)
        ax[i].set_xticklabels(labels = x_dates, rotation = 60, ha = 'right')
        ax[i].set_xlabel("Date of Payment", labelpad = 12)
        ax[i].set_ylabel(ylab, labelpad = 12)
        ax[i].set_title(title, fontsize = 14)
    
    plt.subplots_adjust(hspace = 0.4)
    return plt.show()
```

Using the rules listed below, I calculated the commission contract value (CCV) and total commission payment for each opportunity. 

* CCV is equal to 12 months of software revenue, 12 months of payments revenue, and 50% of hardware and services revenue
* Total commission for SMB restaurants is 9% of CCV, and 7.5% of CCV for MM restaurants.
* First commission payment is 50% of the total commission and is distributed when the customer books.
* Second commission payment is 50% of the total commission and is distributed when the customer goes live.
* If customer churns before they go live, the first half of the rep's commission payment is clawed back.
* If customer puts down a deposit, the rep gets a deposit bonus equal to one month's software revenue which is paid out with the first commission payment. If the customer churns before going life, this deposit bonus is also clawed back.

I started by subsetting the data to view any opportunity that closed in June of 2019 or went live in 2019 for Johnny Cash, Tina Turner, and Steven Tyler only.

```python
from datetime import datetime as dt

# names of interest
persons = ['Johnny Cash', 'Tina Turner', 'Steven Tyler']

# filter commissions data for June 2019
start = dt.strptime('2019-06-01', '%Y-%m-%d')
end = dt.strptime('2019-06-30', '%Y-%m-%d') 
c1 = sales['close_date'] >= start
c2 = sales['close_date'] <= end
c3 = sales['live_date'] >= start
c4 = sales['live_date'] <= end
c5 = sales['opp_owner'].isin(persons)
june = sales[((c1 & c2) | (c3 & c4)) & c5].copy()

# view opportunities by person
for person in persons:
    print(f'\nOpportunities owned by {person} in June 2019\n')
    display(june[june['opp_owner'] == person])
```

I next calculated the commission contract value for each of the representative's opportunities in June 2019 based on the criteria mentioned above. I provided a visualization of the breakdown in revenue contributions for each opportunities' CCV.

```python
import numpy as np

def calc_ccv(df):
    '''
    Calculate CCV for all opportunities in data frame
    '''
    return 12 * (df['saas_rev'] + df['payment_rev']) + .5 * (df['hardware_val'] + df['service_val'])

def calc_commission(df):
    '''
    Calculate representative commission from CCV for all opportunities in data frame
    '''
    ccv = calc_ccv(df)
    if df['restaurant_type'] == 'SMB': return ccv * .09
    else: return ccv * .075

def pay_val(df):
    '''
    Determine net and gross value of commission payments 1 and 2 for all opportunities in data frame
    '''
    commission = calc_commission(df) # total commission
    recovered = commission / 2 * -1 if df['churn_date'] < df['live_date'] or (df['live_date'] is pd.NaT and df['churn_date'] is not pd.NaT) else 0 # "clawback" amount
    kicker = df['saas_rev'] if df['deposit_received'] == True and recovered == 0 else 0 # deposit kicker
    net_p1 = commission / 2 # net payment 1 excluding any clawbacks
    gross_p1 = net_p1 + kicker + recovered # gross payment 1 including clawbacks and deposit kicker
    gross_p2 = commission / 2 # gross payment 2
    return commission, recovered, kicker, net_p1, gross_p1, gross_p2

# calculate ccv for each opportunity
june['ccv'] = june.apply(calc_ccv, axis = 1)
display(june.loc[june['opp_owner'].isin(persons), ['opp_name','restaurant_type','opp_owner','saas_rev','payment_rev','hardware_val','service_val','ccv']].sort_values(by = ['opp_owner']))

# plot distribution of ccv
plot_ccv_bars(june, persons)
```

I next looked at the total commission the representatives received (excluding clawbacks and deposit kickers) from each opportunity based on the opportunity's restuarant type.

```python
# calculate gross commission payments for each opportunity's ccv
for i, name in enumerate(['total_commission', 'recovered','kicker','net_p1', 'gross_p1', 'gross_p2']):
    june[name] = [vals[i] for vals in june.apply(pay_val, axis=1)]

display(june.loc[june['opp_owner'].isin(persons), ['opp_name','restaurant_type','opp_owner','ccv','total_commission']].sort_values(by = ['opp_owner']))
plot_ccv_pct(june, persons)
```

The last thing I looked at was the total payments a sales rep can expect to receive for the month of June 2019. I displayed these results in a cumulative view by day of month.

```python
# create new column to store payment date
for i, name in enumerate(['pmt_dt','pmt_amt']):
    june[name] = [vals[i] for vals in june.apply(pay_details, axis = 1)]
    
# sum payments by user and payment date
june_grp = june[['opp_owner','pmt_dt','pmt_amt']].groupby(['opp_owner', 'pmt_dt']).sum().reset_index().sort_values(by=['opp_owner','pmt_dt'])#level = 'payment_date')
june_grp['pmt_cumsum'] = june_grp.groupby('opp_owner')['pmt_amt'].transform(pd.Series.cumsum)

# fill in missing June dates per each owner
dates = pd.date_range(start='06/01/2019', end = '06/30/2019')
idx = pd.MultiIndex.from_product((dates, persons), names=['pmt_dt','opp_owner'])
june_grp = june_grp.set_index(['pmt_dt', 'opp_owner']).reindex(idx, fill_value = 0).reset_index().sort_values(by=['opp_owner','pmt_dt'])

print('\nTotal commission payouts for the month of June 2019:\n')
display(june_grp.groupby('opp_owner')['pmt_amt'].sum())
print()
plot_pmtbyDay(june_grp, persons)
```
